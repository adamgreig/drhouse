find beats per minute for a song, and find where those beats occur.

begin by loading up scipy stuff

```python
from scipy.signal import iirfilter, firwin, lfilter, fftconvolve
from scipy.io import wavfile
from multiprocessing import Pool
from functools import partial
```

read some wav file, take the left channel, cut out 20s of audio or so

```python
wav = wavfile.read("objection.wav")
sr = wav[0]
source = wav[1][:, 0]
source = source[30.4*sr:40.4*sr]
```

make a filter bank of lpf/bpf/hpf filters.

at first this used the iirfilter but that becomes unstable very quickly with high order so use fir instead.

```python
bank_freqs = [(None, 200), (200, 400), (400, 800), (800, 1600), (1600, 3200), (3200, None)]

def make_filter(freqs):
    if not freqs[0]:
        btype = 'lowpass'
        freqs = freqs[1] / (sr / 2)
    elif not freqs[1]:
        btype = 'highpass'
        freqs = freqs[0] / (sr / 2)
    else:
        btype = 'bandpass'
        freqs = freqs[0] / (sr / 2), freqs[1] / (sr / 2)
    return iirfilter(3, freqs, rp=3, rs=40, btype=btype, ftype='ellip')

def make_fir_filter(freqs):
    if not freqs[0]:
        pass_zero = True
        freqs = freqs[1]
    elif not freqs[1]:
        pass_zero = False
        freqs = freqs[0]
    else:
        pass_zero = False
    return firwin(127, freqs, pass_zero=pass_zero, nyq=sr/2), 1

with Pool() as pool:
    filterbank = pool.map(make_fir_filter, bank_freqs)
```

filter the source into six filtered components

```python
def apply_source_filter(fil):
    return lfilter(fil[0], fil[1], source)

with Pool() as pool:
    source_bands = pool.map(apply_source_filter, filterbank)
    rectified_bands = pool.map(np.abs, source_bands)
```

now convolve with a half hanning window to extract the signal envelope

```python
window = np.hanning(sr * 0.2 * 2)
window = window[window.size / 2 :]
f = partial(fftconvolve, in2=window, mode='same')
with Pool() as pool:
    envelopes = pool.map(f, rectified_bands)
```

to demonstrate that the envelope contains all you need to recover the beat, use it to modulate white noise sent through the same filter bank, then we can listen to that noise and see if we can still hear the beat

```python
noise = np.random.randn(source.size)
def apply_noise_filter(fil):
    return lfilter(fil[0], fil[1], noise)

with Pool() as pool:
    noise_bands = pool.map(apply_noise_filter, filterbank)
```

```python
modulated_noise_bands = np.multiply(np.array(envelopes), np.array(noise_bands))
```

```python
modulated_noise_bands = []
for i in range(len(envelopes)):
    modulated_noise_bands.append((envelopes[i] ** 2) * noise_bands[i])
```

```python
output = np.sum(modulated_noise_bands, axis=0)
output /= np.max(output)
```

write out the wavs

```python
wavfile.write("output.wav", sr, output)
```

we can still hear the beat! so the envelopes contain all we need. let's see 'em

```python
plt.figure(figsize=(18, 12))
for i in range(6):
    plt.subplot(2, 3, i)
    plt.plot(envelopes[i])
```

(this is more or less obviously beaty in different songs)

okay so we'll follow http://www.iro.umontreal.ca/~pift6080/H09/documents/papers/scheirer_jasa.pdf and differentiate and then half wave rectify the envelopes and then downsample to 200Hz to make computation easier.

```python
def half_wave_rectify(x):
    return x.clip(0)

dssr = 200

def downsample(x):
    return x[::sr/dssr]

with Pool() as pool:
    envelopes_diff = pool.map(np.diff, envelopes)
    envelopes_rect = pool.map(half_wave_rectify, envelopes_diff)
    envelopes_ds = pool.map(downsample, envelopes_rect)
```

the Scheirer paper uses a bank of resonant filters to try and find the bpm, but let's see if a simple fft will work

```python
def rfft(x):
    return np.abs(np.fft.rfft(np.pad(x, (0, x.size), mode='constant')))

with Pool() as pool:
    rect_freqs = pool.map(rfft, envelopes_ds)
```

```python
freqs = np.fft.rfftfreq(envelopes_ds[0].size * 2, 1/dssr)
plt.figure(figsize=(18, 12))
for i in range(6):
    plt.subplot(2, 3, i)
    plt.plot(freqs[5:100], rect_freqs[i][5:100])
    plt.grid()
```

add them all together (yes you can add the envelopes then do one fft, will do that later)
and use it to find the beats

```python
fftsum = np.sum(rect_freqs, axis=0)
plt.figure(figsize=(18, 5))
plt.plot(freqs[5:100], fftsum[5:100])
#plt.xticks(np.arange(0, 5, .1))
plt.grid()
```

```python
envsum = np.sum(envelopes_ds, axis=0)
fftsum2 = np.abs(np.fft.rfft(np.pad(envsum, (0, envsum.size), mode='constant')))
plt.figure(figsize=(18, 5))
plt.plot(freqs[5:100], fftsum[5:100])
plt.plot(freqs[5:100], fftsum2[5:100])
plt.grid()
```

```python
plt.figure(figsize=(18, 5))
plt.plot(envsum[:5*dssr])
plt.plot(source[::sr/dssr][:5*dssr])
```

```python
n = envelopes_ds[0].size * 2
d = 1/dssr

def freq_to_index(freq):
    return int(freq * ((n * d) / 2))
```

we expect the bpm to be the lowest frequency with lots of overtones that has peaks in the spectrum. for now, try every bpm in a big range, and find the one with the strongest fundemental plus overtones

```python
max_evidence = 0
best_bpm = None
evidences = []
bpms = np.arange(50, 210, 0.1)
for bpm in bpms:
    bps = bpm / 60.0
    tones = np.array([freq_to_index(bps * i) for i in (1, 2, 4, 8)])
    evidence = np.sum(fftsum[tones])
    evidence += np.sum(fftsum[tones - 1])
    evidence += np.sum(fftsum[tones + 1])
    if evidence > max_evidence:
        max_evidence = evidence
        best_bpm = bpm
    evidences.append(evidence)
print(best_bpm)
plt.figure(figsize=(18, 5))
plt.plot(bpms, evidences)
plt.grid()
```

convincing! perhaps anyway. this is very often the correct BPM for the song.

now try to figure out when the beat happened, correlate an impulse train with the envelopes

```python
best_bps = best_bpm / 60.0
period = 1 / best_bps

impulses = np.zeros_like(envsum).astype(np.float)
beat_locations = np.arange(0, envsum.size, period * dssr)
for beat in beat_locations:
    impulses[beat:beat+1] = 1


cor = fftconvolve(envsum, impulses, mode='same')
shift = np.argmax(cor) / dssr
#shift += period / 2
shift %= period

plt.figure(figsize=(18, 4))
plt.plot(cor[:period * dssr * 5])
plt.axvline(shift * dssr, color='g')
plt.axvline(shift * dssr + period * dssr, color='g')
plt.axvline(shift * dssr + period * dssr * 2, color='g')

print(shift)
print(shift * dssr)
```

```python
plt.figure(figsize=(18, 6))
plt.plot(envsum[:4*dssr] / np.max(envsum))
plt.plot(np.pad(impulses, (shift*dssr, 0), mode='constant')[:4*dssr])
```

```python
click = 0.3 * np.sin(2*np.pi*1000.0*np.arange(0, 0.02, 1/sr))
click += 0.3 * np.sin(2*np.pi*2000.0*np.arange(0, 0.02, 1/sr))
click += 0.3 * np.sin(2*np.pi*4000.0*np.arange(0, 0.02, 1/sr))


clicktrack = np.zeros_like(source).astype(np.float)

beats = np.arange(shift*sr, clicktrack.size - click.size, period * sr)
for beat in beats:
    clicktrack[beat:beat + click.size] += click

sourceflt = source.copy().astype(np.float) / np.max(source)
clicks = sourceflt + clicktrack

wavfile.write("clicks.wav", sr, clicks)

plt.figure(figsize=(18, 6))
plt.plot(clicktrack[:3*sr])
plt.plot(sourceflt[:3*sr])
```
