```python
from scipy.io import wavfile
from scipy.signal import iirfilter, lfilter
```

```python
wav = wavfile.read("./noisyclick.wav")
sr = wav[0]
source = wav[1]
x = source[10*sr : 20*sr]
plt.plot(x)
```

```python
spec = np.fft.rfft(x)
freqs = np.fft.rfftfreq(x.size, 1/sr)
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.plot(freqs, spec)
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(freqs[:50], spec[:50])
plt.grid()
```

```python
cutoff = 15.0 / (sr / 2.0)
fil_b, fil_a = iirfilter(5, cutoff, btype='lowpass', ftype='butter')
```

```python
y = lfilter(fil_b, fil_a, x)
plt.plot(y)
```

```python
win = np.hamming(y.size)
z = np.pad(y * win, (0, 20*r), 'constant')
f = np.log(np.abs(np.fft.rfft(z)))
w = np.fft.rfftfreq(z.size, 1/r)
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.plot(w, f)
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(w[:100], f[:100])
plt.grid()
```
